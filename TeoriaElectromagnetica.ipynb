Las funciones de la bilbioteca numpy que se emplearán son:
- Para crear arreglos:
 - np.array([valores])
 - np.arange(start, stop, step)
 - Para conocer el tipo de arreglo np.shape
- Producto punto: np.dot(a, b), también empleado para multiplicar matrices
- Matriz transpuesta: np.transpose()
- Producto cruz: np.cross(a, b)

Ejemplos:

```python
import numpy as np
a = np.array([3, 4, -2])
b = np.array([-1, 1, 3])

print('a es: ', a.shape, 'b es: ', b.shape)
print('Producto punto', np.dot(a,b))
print('Producto cruz', np.cross(a,b))
```

## 2. Gráficos

Gráficos con la quiver de matplotlib.

```python
#Establecer espacio para graficar
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

fig, ax = plt.subplots(2, 2, figsize=(6, 6))
ax[0][0].quiver(x,y,1,1)
ax[0][1].quiver(x,y,1,-1)
ax[1][0].quiver(x,y,-1,1)
ax[1][1].quiver(x,y,-1,-1)
fig.show()
```

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))
u = x/np.sqrt(x**2 + y**2)
v = y/np.sqrt(x**2 + y**2)
plt.quiver(x,y,u,v)
plt.show()
```

Gráfica con streamplot().

```python
# Meshgrid
X,Y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))
 
# Assign vector directions
Ex = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
Ey = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2)
 
fig, ax = plt.subplots(figsize=(6, 6))
ax.streamplot(X,Y,Ex,Ey, density=1.0, linewidth=0.5, color='#A23BEC')
ax.plot(-1,0,'-or')
ax.plot(1,0,'-og')
ax.grid()
fig.show()
```
